```
#LOAD AND PREPARE DATA
nnr = NNRegressor()
nnr.read_train_data("https://raw.githubusercontent.com/motokazmin/kaggle/ML_Boot_Camp/Performance_evaluation/data/x_train.csv",
                    "https://raw.githubusercontent.com/motokazmin/kaggle/ML_Boot_Camp/Performance_evaluation/data/y_train.csv",
                    drop_cat = True, test_size = 0.2, target='time')

nnr.prepare_train_data()
```

```
#VALIDATE AND STUDY DATA
rru.validate(nnr)
rru.show_pca_explained_variances(nnr)
```

```
#MAKE REGRESSOR
nnr.prepare_regressor("ExtraTreesRegressor", n_estimators = 100, min_samples_leaf = 1, min_samples_split = 2, criterion='mae')
nnr.no_tune_model()
```
